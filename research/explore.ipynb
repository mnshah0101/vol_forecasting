```python
import pandas as pd
```

```python
df = pd.read_parquet('/tmp/BTC_USDT.parquet')
```

```python

df["received_time"] = pd.to_datetime(df["received_time"])
df = df.sort_values("received_time").reset_index(drop=True)

gaps = df["received_time"].diff()
holes = df[gaps > pd.Timedelta(minutes=30)].copy()
holes["gap_start"] = df["received_time"].shift(1)[holes.index]
holes["gap_duration"] = gaps[holes.index]

print(holes[["gap_start", "received_time", "gap_duration"]].rename(columns={"received_time": "gap_end"}))
```
